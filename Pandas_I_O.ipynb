### **Banco de Dados**

```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
#Criando um banco de Local nativo da biblioteca
engine = create_engine('sqlite:///:memory:')
```

```python
url_bd = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'

df_clientes = pd.read_csv(url_bd)
```

```python
#Inserindo dados em uma tabela no banco
df_clientes.to_sql('clientes', engine, if_exists='replace') # Change the if_exists parameter
```

```python
#Verificando
inspector = inspect(engine)
inspector.get_table_names()
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
empregados = pd.read_sql(query, engine)
empregados.head()
```

```python
#Salvando consulta como Tabela

empregados.to_sql('empregados',con=engine, if_exists='replace',index=False)
```

```python
pd.read_sql_table('empregados',engine)
```

```python
from sqlalchemy import text

query_delete = text('DELETE FROM clientes WHERE ID_Cliente=5008805')

# Conectando ao banco de dados e executando a consulta
with engine.connect() as conn:
    conn.execute(query_delete)
```

```python
pd.read_sql_table('empregados',engine)
```

```python
from sqlalchemy import create_engine, text

# Supondo que você já tenha um engine configurado
# engine = create_engine('sua_string_de_conexão')

query_update = text("UPDATE clientes SET Grau_escolaridade='Ensino superior' WHERE ID_Cliente=:id_cliente")

# Conectando ao banco de dados e executando a consulta
with engine.connect() as conn:
    conn.execute(query_update, {'id_cliente': 5008806})
    conn.commit()
```
