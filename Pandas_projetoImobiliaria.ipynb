# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
mediaAlugeis = dados.groupby('Tipo').mean(numeric_only=True) #agrupando os dados e pegando a media pelo tipo
mediaAlugeis
```

```python
mediaAlugeis = dados.groupby('Tipo')['Valor'].mean(numeric_only=True)
display(mediaAlugeis)
```

```python
mediaAlugeis = dados.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor') #transformando em um DataFrame e ordenando
display(mediaAlugeis)
```

```python
df_media_preco_tipo = dados.groupby('Tipo')[['Valor']].mean(numeric_only=True).sort_values('Valor')
df_media_preco_tipo.plot(kind='barh', figsize=(14, 10), color='blue')

```
